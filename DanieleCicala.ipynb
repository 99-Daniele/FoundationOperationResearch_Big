## Biogas plants location

An association of $n$ farmers wants to open $p$ plants to produce energy from biogas. 
Each plant will be opened at a farm of a member of the association and will be powered with corn chopping purchased from the farm itself or from other neighboring farms.

Each farm $i$ can provide at most $c_i$ tons of corn chopping, with a percentage of dry matter $a_i$. As you may know, dry matter is the key component of corn chopping used for biogas production. In order to maintain the quality of produced energy, each plant must burn a mixture of corn chopping with a percentage of dry matter between $k_{min}$ and $k_{max}$. 

At most one plant can be located in each farm, and every farm can sell its corn chopping to one and only one plant.

Each farm $i$ is located at coordinates $x_i$ and $y_i$, representing respectively its latitude and longitude, and the cost of moving corn chopping from a farm $i$ to a farm $j$ is proportional to the euclidean distance between the two farms (it does not depend on the actual quantity moved, since the trucks used for this transportations are sufficiently big). 

Under such conditions, every plant produces $Q$ kWh of energy per ton of corn chopping burned. The energy produced by each plant will be fed into the national electricity system, at a unitary price of $b$ (€/kWh). Moreover, due to state regulations, each plant must not produce more than $M$ kWh of energy.

You must locate $p$ plants among the available farms and assign the farms that will supply each plant, with the goal of maximizing the total revenues of the association.

### Sets
*   $I$ = set of farms

### Parameters
*   $n$ = number of farms   
*   $p$ = number of plants to locate
*   $b$ = revenue per unit of energy (€/kWh)
*   $M$ = max energy production (kWh)
*   $Q$ = energy produced by a ton of corn chopping (kWh/t)
*   $k_{min} (k_{max})$ = min (max) percentage of dry matter for fermentation
*   $a_i$ = percentage of dry matter in chopping from farm $i \in I$
*   $c_i$ = tons of corn chopping available for each $i \in I$ (t)
*   $x_i, y_i$ = coordinates of farm $i \in I$

```python
# Return the euclidean distance between two points.
def euDistance(p1, p2):
  euclideanDistance = ((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2) ** 0.5
  return euclideanDistance

# Return the second elem of list. Used for sorting list of list based on second element.
def takeSecond(elem):
  return elem[1]

# Return plant from plants with given id.
def returnPlant(plants, id):
  for plant in plants:
    if plant[0] == id:
      return plant

# Find the kth plant with lower distance from given farms. 
def findNearestPlant(farms, plantPlannings, distances, k):
  plants = [[plant[0], sum(distances[farm][plant[0]] for farm in farms)] for plant in plantPlannings]
  plants.sort(key=takeSecond)
  return plants[k][0]

# Find the nearest empty plant. If there isn't any empty plant find kth plant with lower distance
def findNearestEmptyPlant(farms, plantPlannings, distances, k):
  p = len(plantPlannings)
  for j in range(p):
    plant = returnPlant(plantPlannings, findNearestPlant(farms, plantPlannings, distances, j))
    if plant[1] == 0:
      return plant[0]
  return findNearestPlant(farms, plantPlannings, distances, k)       

# Fill each plant with at least one farm supplie without exceeding max corn and dry matter percentage limits.   
def populatePlants(farms, plants, kmin, kmax, M, distances):
  p = len(plants)
  n = len(farms)
  # In order to decide which farm goes to which plant, we create two sublist: candidates are the p best farms based on how much they can provide and others are the remaining ones.
  candidates = farms[0:p]
  others = farms[p:n]
  while len(candidates) > 0:
    farm = candidates[0]
    # plan is one possible combination of chosen candidates and other farms such that the total amount of corn provided is less than maximum and the percentage of dry matter is inside the limits.
    # plan is also a combination with 0 waste which means that selected farms provides all available corn without reductions and the solution with less farms involved.
    plan = zeroWastePlan(farm[1], farm[2], kmin, kmax, M, others)
    t = len(plan)
    if t == 0 or plan[0].x != None:
      # chosenFarms is the list of index of planned farms. farmsParameters is the list of farm corn and dry matter for each chosen farms.
      chosenFarms = [farm[0]]
      farmsParameters = [[farm[1], farm[2]]]
      for k in range(t):
        if plan[k].x == 1:
          chosenFarms.append(others[k][0])
          farmsParameters.append([others[k][1], others[k][2]]) 
      totalCorn = sum(farmParam[0] for farmParam in farmsParameters)
      dryMatterPercentage = sum(farmParam[0] * farmParam[1] for farmParam in farmsParameters) / totalCorn
      for j in range(p):
        # After finding one solution we find the nearest plant.
        plant = returnPlant(plants, findNearestPlant(chosenFarms, plants, distances, j))
        # If the nearest plant was previously provided by other farms we find the 2nd nearest, then the 3rd one and so on.
        if plant[1] == 0:
          # If plant has not yet provided by any farm we select it and fill with the amount of corn, the dry matter percentage and the total shipping cost.
          shippingCost = sum(distances[chosenFarm][plant[0]] for chosenFarm in chosenFarms)
          plant[1] = totalCorn
          plant[2] = dryMatterPercentage
          plant[3] = shippingCost
          # After that we remove chosen farms from the available ones to avoid one farm providing more plants. We also delete from candidate and others.
          for i in range(t):
            if plan[t - i - 1].x == 1:
              farms.pop(farms.index(others[t - i - 1]))
              others.pop(others.index(others[t - i - 1]))   
          farms.pop(farms.index(candidates[0]))
          candidates.pop(0)    
          break
    # If plan is None then we didn't find any solution with 0 waste. We move the candidate to the end of others list and the head of others list to candidates.      
    else:
      # In case the last candidate is lower than the first of others than we are entering in a loop. To avoid this we return directly.
      if candidates[-1][1] < others[0][1]:
        return
      candidates.append(others[0])
      others.append(candidates[0])
      candidates.pop(0)
      others.pop(0)

# Return a combination of farms such that the total amount of corn and dry matter percentage respects the limits and all selected farms send all available corn without wasting it.
# The combination is the one with less farms involved.
def zeroWastePlan(corn, dryMatter, kmin, kmax, M, farms):
  n = len(farms)
  if n == 0:
    return [] 
  m = mip.Model()
  # x is the amount of corn provided by each farm.
  x = [m.add_var() for i in range(n)]
  # y is an auxiliary array of variables. When x[i] > 0 then y[i] == 1. 
  y = [m.add_var(var_type=mip.BINARY) for i in range(n)]
  # waste is the difference between how much corn can be provided and how much we effectively provide if we select those farms.
  waste = mip.xsum(farms[i][1] * y[i] - x[i] for i in range(n))
  # farmCount is the amount of farms involved.
  farmCount = mip.xsum(y[i] for i in range(n))
  # maxCorn is the maximum amount of corn that can be provided by each farm.
  maxCorn = [farms[i][1] for i in range(n)]
  # newCorn is how much corn is provided from other farms.
  newCorn = mip.xsum(x[i] for i in range(n))
  # newDryMatter is the amount of dry matter provided.
  newDryMatter = mip.xsum(x[i] * farms[i][2] for i in range(n))
  totalCorn = corn + newCorn
  totalDryMatter = corn * dryMatter + newDryMatter
  maxCornConstr = [m.add_constr(x[i] <= maxCorn[i] * y[i]) for i in range(n)]
  MCornConstr = m.add_constr(totalCorn <= (M - 0.000001))
  minDryMatterConstr = m.add_constr(totalDryMatter >= totalCorn * (kmin + 0.000001))
  maxDryMatterConstr = m.add_constr(totalDryMatter <= totalCorn * (kmax - 0.000001))
  # we force waste to 0 and minimize farmCount.
  m.add_constr(waste == 0)
  m.objective = mip.minimize(farmCount)
  m.optimize()
  return y 

# Decide to which plant given farms will provide its corn. 
def placeFarm(farm, plants, distances, kmin, kmax, M, farms, Q, b):
  otherFarms = farms[1:]
  p = len(plants)
  for j in range(p):
    # We find the nearest plant from given farm and find out if exist a combination with 0 waste. If not we move to the 2nd nearest, then the 3rd ones and so on.
    plant_idx = findNearestEmptyPlant([farm[0]], plants, distances, j)
    plant = returnPlant(plants, plant_idx)
    newCorn = plant[1] + farm[1]
    newDryMatter = (plant[1] * plant[2] + farm[1] * farm[2]) / newCorn
    plan = zeroWastePlan(newCorn, newDryMatter, kmin, kmax, M, otherFarms)
    if plan[0].x != None:
      # If we find a feasible solution we calculate the new total amount of corn, the new dry matter percentage, the new shipping cost and change selected plants parameters.
      t = len(plan)
      totalCorn = newCorn + mip.xsum(otherFarms[k][1] * plan[k] for k in range(t))
      dryMatterPercentage = (mip.xsum(otherFarms[k][1] * otherFarms[k][2] * plan[k] for k in range(t)) + (plant[1] * plant[2] + farm[1] * farm[2])) / totalCorn.x
      shippingCost = distances[farm[0]][plant_idx] + mip.xsum(distances[otherFarms[k][0]][plant_idx] * plan[k] for k in range(t))
      plant[1] = totalCorn.x
      plant[2] = dryMatterPercentage.x
      plant[3] = plant[3] + shippingCost.x
      # We remove all farms involved.
      for i in range(t):
        if plan[t - i - 1].x == 1:
          farms.pop(farms.index(otherFarms[t - i - 1]))
      farms.pop(farms.index(farm))
      return -1
  # If there isn't any plant which has a solution with 0 waste we move it to the end of farms and return the index of that farm.
  discarded = farms[0][0]    
  farms.append(farms[0])
  farms.pop(0)
  return discarded   

# Given the last farm, find plant with best gain (biogas produced - distance) which did not exceed the limits.
def findLastPlant(farm, plants, kmin, kmax, M, distances, Q, b):
  max = 0
  newCorn = 0
  newDryMatter = 0
  newShippingCost = 0
  chosenPlant = []
  p = len(plants)
  end = False
  # For each plant we calculates how much is the gain if we respect the limits. The one with the best gain will be chosen.
  for plant in plants: 
    # If there is a plant which still does not respect the limits we select it. Actually by construction this should never happen.     
    if (plant[2] < kmin or plant[2] > kmax) and plant[1] > 0:
      end = True
    shippingCost = distances[farm[0]][plant[0]]  
    m = mip.Model()
    x = m.add_var()
    dryMatter = plant[1] * plant[2] + x * farm[2]
    m.add_constr(x + plant[1] <= (M - 0.000001))
    m.add_constr(x <= farm[1])
    m.add_constr(dryMatter >= (kmin + 0.000001) * (x + plant[1]))
    m.add_constr(dryMatter <= (kmax - 0.000001) * (x + plant[1]))
    gain = x * Q * b - shippingCost
    m.objective = mip.maximize(gain)
    m.optimize()
    if gain.x != None:
      # If we have found a plant that don't satisfies the limits or a gain which is greater than others we save that parameters.
      if end == True or gain.x > max:
        newCorn = x.x
        newDryMatter = dryMatter.x / (x.x + plant[1])
        newShippingCost = shippingCost
        chosenPlant = plant
        if end == True:
          break
        elif gain.x > max:
          max = gain.x
  # If there is actually a chosen plant we change its parameters.      
  if chosenPlant != []:               
    chosenPlant[1] = chosenPlant[1] + newCorn
    chosenPlant[2] = newDryMatter     
    chosenPlant[3] = chosenPlant[3] + newShippingCost 

# Given farm, find which is the combination with less waste.
def findBestPlant(farm, plants, kmin, kmax, M, farms, distances):
  max = 99999
  corn = 0
  dryMatter = 0
  shippingCost = 0
  chosenPlant = []
  wasteY = []
  p = len(plants)
  n = len(farms)
  for plant in plants:
    m = mip.Model()
    x = [m.add_var() for i in range(n)]
    y = [m.add_var(var_type=mip.BINARY) for i in range(n)]
    waste = mip.xsum(farms[i][1] * y[i] - x[i] for i in range(n))
    maxCorn = [farms[i][1] for i in range(n)]
    newCorn = mip.xsum(x[i] for i in range(n))
    newDryMatter = mip.xsum(x[i] * farms[i][2] for i in range(n))
    totalCorn = plant[1] + newCorn
    totalDryMatter = plant[1] * plant[2] + newDryMatter
    maxCornConstr = [m.add_constr(x[i] <= maxCorn[i] * y[i]) for i in range(n)]
    MCornConstr = m.add_constr(totalCorn <= (M - 0.000001))
    minDryMatterConstr = m.add_constr(totalDryMatter >= totalCorn * (kmin + 0.000001))
    maxDryMatterConstr = m.add_constr(totalDryMatter <= totalCorn * (kmax - 0.0000001))
    # We force the first farm (which is the given one) to provide more than 0 corn.
    m.add_constr(y[0] == 1)
    m.objective = mip.minimize(waste)
    m.optimize()
    # If we have found waste which is less than others we save that parameters.
    if waste.x != None and waste.x < max:
      max = waste.x
      corn = totalCorn.x
      dryMatter = totalDryMatter.x / totalCorn.x
      chosenPlant = plant
      wasteY = y 
  # If we have found at least one solution we change chosen plant parameters and delete chosen farms.    
  if wasteY != []:    
    shippingCost = mip.xsum(wasteY[i] * distances[farms[i][0]][plant[0]] for i in range(n))         
    chosenPlant[1] = corn
    chosenPlant[2] = dryMatter     
    chosenPlant[3] = chosenPlant[3] + shippingCost.x
    for i in range(n):
      if wasteY[n - i - 1].x == 1:
          farms.pop(farms.index(farms[n - i - 1]))
  # If no solutions has been found we still remove the farm from available ones.     
  else:
    farms.pop(0)

def solve(instance):

  # Get all parameters.
  M = instance['M']
  Q = instance['Q']
  a = instance['a']
  b = instance['b']
  c = instance['c']
  kmax = instance['kmax']
  kmin = instance['kmin']
  n = instance['n']
  p = instance['p']
  points = instance['points']
  MAX = M / Q

  # Calculate the distances from each farm to each other.
  distances = [[euDistance(points[i], points[j]) for j in range(n)] for i in range(n)]

  # For each farm calculate the average distance from other farms. Select the p lower ones where plants will be built.
  selectedPlants = []  
  for i in range(n):
    distance = sum(distances[i][j] for j in range(n)) / n
    if len(selectedPlants) < p or distance < selectedPlants[0][1]:
      selectedPlants.append([i, distance])
      selectedPlants.sort(key=takeSecond, reverse=True)
      if len(selectedPlants) > p:
        selectedPlants.pop(0)

  # plantPlannings summarize for each plant their id (farm id), the amount of corn that is now burned, the current percentage of dry matter,
  # and the total costs of shipping from other farms.
  plantPlannings = [[plant[0], 0.0, 0.0, 0.0] for plant in selectedPlants]

  # availableFarms is the list of farm which have not yet plan where to send their corn. Summarize their id, how much corn can provide and its dry matter percentage.
  # This list is at the beginning ordered by the amount of corn descending
  availableFarms = [[i, c[i], a[i]] for i in range(n)]
  availableFarms.sort(key=takeSecond, reverse=True)

  # Fill each plant with at least one farm supplie without exceeding max corn and dry matter percentage limits. 
  populatePlants(availableFarms, plantPlannings, kmin, kmax, MAX, distances)
  
  discarded = -1
  loop = False
  # Plan where to sell their corn for each remaining farm.
  while len(availableFarms) > 0:
    # If there is only one remaining farm we find the best one based on how much corn can be provided and how much it's the distance from the farm.
    if len(availableFarms) == 1:
      findLastPlant(availableFarms[0], plantPlannings, kmin, kmax, MAX, distances, Q, b)
      # Farm is then pop out whether a solution has been found or not. 
      availableFarms.pop(0)
      break
    # If availableFarms[0][0] == discarded then the first farm moved to last farm is now the head of the list again. So we have scan all the list to come back to the beggining.
    # To avoid entering in a loop from now on we will search the best plant for that farm. In case there is still no solution the farm is removed anyway.       
    elif availableFarms[0][0] == discarded or loop == True:
      findBestPlant(availableFarms[0], plantPlannings, kmin, kmax, MAX, availableFarms, distances)
      loop = True
    else:
      dis = placeFarm(availableFarms[0], plantPlannings, distances, kmin, kmax, MAX, availableFarms, Q, b)       
      # If dis != -1 then the farm has been moved to the end of the list. If discarded == -1 then this is the first time one farm is moved to the end. We saved it in orded to know when we find that farm again.
      if dis != -1 and discarded == -1:
        discarded = dis

  # For each plant calculates how much biogas its procuded and subtract shipping costs. The sum of it it's the total revenue.
  totalRevenues = sum(plant[1] * Q * b - plant[3] for plant in plantPlannings)

  # This is printed only if at the end the solution doesn't respect the limits: one plant burned too much corn and/or dry percentage of one plant exceed the limit
  for j in range(p):
    max_ok = plantPlannings[j][1] < MAX
    dry_matter_ok = (plantPlannings[j][2] >= kmin and plantPlannings[j][2] <= kmax) or plantPlannings[j][2] == 0
    if max_ok == True and dry_matter_ok == False:
      print(j, "{}".format(plantPlannings[j][1]), "{}".format(MAX), max_ok, "", "{:.7f}".format(plantPlannings[j][2]), "{:.7f}".format(kmin), "{:.7f}".format(kmax), "{:.16f}".format(plantPlannings[j][2] - kmax))
    elif max_ok == False:
      print(j, "{}".format(plantPlannings[j][1]), "{}".format(MAX), max_ok, "{:.7f}".format(plantPlannings[j][2]), "{:.7f}".format(kmin), "{:.7f}".format(kmax), dry_matter_ok)

  return totalRevenues
```
